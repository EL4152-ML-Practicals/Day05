```python
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv('FSdata.csv')
print("Dataset: \n", data)
```

```python
# Extracting attributes and target variable
concepts = np.array(data)[:, :-1]
print("Concepts (Attributes): \n", concepts)
```

```python
target = np.array(data)[:, -1]
print("Target Variable: \n", target)

# Find-S Algorithm Implementation
def train(con, tar):
    for i , val in enumerate(tar):
        if val.lower() == 'yes': # Case-insensitive match for 'yes'
            specific_h = con[i].copy()
            break

    for i, val in enumerate(con):
        if tar[i].lower() == 'yes':
            for x in range(len(specific_h)):
                if val[x] != specific_h[x]:
                    specific_h[x] = '?'
    return specific_h

# Final Hypothesis
final_hypothesis = train(concepts, target)
print("Final Hypothesis: \n", final_hypothesis)
```
